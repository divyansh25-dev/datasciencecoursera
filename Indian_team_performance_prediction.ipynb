```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/ContinousDataset.csv')
```

```
df
```

```
size_mapping = {'Margin':{'Winner2ndInning':0,'Winner1stInning':1},
                'Innings_Team1':{'Second':1,'First':0},'Innings_Team2':{'Second':1,'First':0}
                }
df.replace(size_mapping,inplace= True)
df
```

```

def j(dx,variable,top_x_label):
  for label in top_x_label:
    dx[variable+'_'+label]=np.where(dx[variable]==label,1,0)
```

```
df
```

```
top_10 = [x for x in df['Team 1'].value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```
j(df,'Team 1',top_10)   
```

```
j(df,'Team 2',top_10)   
```

```
df
```

```
top_11 = [x for x in df['Host_Country'].value_counts().sort_values(ascending=False).head(11).index]
top_11
```

```
j(df,'Host_Country',top_11)
```

```
df
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Winner'] = le.fit_transform(df['Winner'])
df
```

```
pd.get_dummies(df['Venue_Team1'], prefix='Venue_Team1',drop_first=True)
pd.get_dummies(df['Venue_Team2'], prefix='Venue_Team2',drop_first=True)
```

```
df
```

```
df = pd.concat([df,pd.get_dummies(df['Venue_Team1'], prefix='Venue_Team1')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
df.drop(['Venue_Team1'],axis=1, inplace=True)
```

```
df = pd.concat([df,pd.get_dummies(df['Venue_Team2'], prefix='Venue_Team2')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
df.drop(['Venue_Team2'],axis=1, inplace=True)
```

```
df
```

```
df['res']=df['Winner']
pd.concat([df,df['res']],axis=1)
```

```
df.drop(['Winner'],axis=1, inplace=True)
```

```
df
```

```
x = df.iloc[:,6:-1].values
y = df.iloc[:,-1].values
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x_train, y_train)
```

```
y_pred=regressor.predict(x_test)
```

```
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```
from sklearn.ensemble import RandomForestRegressor
Regressor = RandomForestRegressor(random_state = 0)
Regressor.fit(x_train, y_train)
```

```
Y_pred=Regressor.predict(x_test)
```

```
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, Y_pred)
```
